```python
import pandas as pd
import os
import openpyxl
import matplotlib.pyplot as plt
os.getcwd()
```

```python
filename = "data/cnn_training_output.xlsx"
```

```python
wb = openpyxl.open(filename)
```

```python
for sheet in wb.sheetnames:
    plt.figure(figsize=(10, 6))
    df_cur = pd.read_excel(filename, sheet_name=sheet)
    plt.plot(df_cur["epoch"], df_cur["loss"], label="Training Loss", marker="o")
    plt.plot(df_cur["epoch"], df_cur["val_loss"], label="Validation Loss", marker="o")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(f"{sheet} CNN Training and Validation Loss over Epochs")
    plt.legend()
    plt.grid()
```
